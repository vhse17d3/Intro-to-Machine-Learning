# Pylot API vs Object-Oriented (OO) API
* Pylot API  -> quickly 
* OO API -> advanced

## Pylot API

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),linestyle='dashed',label="sin(x)",color='red')
plt.plot(x,np.cos(x),label="cos(x)",color='blue')
plt.xlabel('Bien X') # to title ca trc
plt.ylabel("F(X)") 
plt.title("Sin(X) and Cos(X)")# to title ca bng
# plt.axis()  gii hn  th 
plt.legend() # xem label ca ng k 
plt.show()
```

# Object-Oriented API

```python
#1. prepare data 
x=[1,2,3,4]
y=[11,22,33,44]
#2. setup plot 
fig,ax =plt.subplots(figsize=(5,5)) # figure size = widh & height of the plot 
#3 plot the data
ax.plot(x,y)
ax.set(title="A simple plot",
      xlabel='x-axis',
      ylabel='y-axis')
```

```python
## OO API 
# prepare New Data 
rng =np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.rand(100)
sizes=500*rng.rand(100)
fig , ax=plt.subplots()
img1=ax.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.3) # alpha l  m , cmap l kiu mu 
fig.colorbar(img1) # to thanh mu ca bng 
```

# Histogram



```python
# prepare data
student_height=np.random.normal(170,10,250)
```

```python
student_height[:10],len(student_height)
```

```python
# pyplot API
plt.hist(student_height,bins=30)
```

```python
# Option 1: OO API
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))
ax1.plot(x,x/2)
ax1.plot(x,x*2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax4.hist(student_height,bins=30)
```

```python
# Option 2:
fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,0].plot(x,x*2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax[1,1].hist(student_height,bins=30)
```

#    # Project data visualizatoin - California Cities

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
file_path = r"C:\Users\sonki\test python\matplotlib v trc quan ha d liu cc thnh ph\data\california_cities.csv"
# c tp bng Pandas
cities = pd.read_csv(file_path)
cities.head()
```

```python
# Extract latd = v  and Longd = kinh 

lat,long=cities['latd'],cities['longd']
population , area = cities['population_total'],cities['area_total_km2']
plt.figure(figsize=(8,6))
# Plot using Pylot API 
plt.scatter(long,lat,
           c=np.log10(population),cmap='viridis',
           s=area,linewidths=0,alpha=0.5)
plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7) # gi tr colorbar
# Create a legend for cities size 
area_range = [50,100,300,500]
for area in area_range: # v ch thch 
    plt.scatter([],[],s=area,c='k',alpha=0.4,
               label=str(area)+'Km$^2$')
plt.legend() # hin th ch thch 
plt.title('California Cities: Population and Area Distribution')
```
