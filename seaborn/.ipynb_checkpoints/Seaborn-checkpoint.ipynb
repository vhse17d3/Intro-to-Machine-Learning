# KDE Plot
* Biểu đồ ước tính mật độ hạt nhân (KDE) là một phương pháp để trực quan hóa sự phân bố các quan sát trong một tập dữ liệu, analagous với biểu đồ.
* KDE đại diện cho dữ liệu bằng cách sử dụng đường cong mật độ xác suất liên tục trong một hoặc nhiều chiều

```python
sns.kdeplot(tips_df['total_bill']) # vẽ biểu đồ đường 
```

# Displot
* là một hàm trong thư viện Seaborn được sử dụng để tạo biểu đồ phân phối (distribution plot). Biểu đồ phân phối giúp bạn hiển thị và hiểu sự phân phối của dữ liệu, thường sử dụng để kiểm tra tính phân phối của dữ liệu và xem các đặc điểm cụ thể như độ lệch (skewness), độ nhọn (kurtosis) của phân phối.

```python
sns.displot(data=tips_df,x="total_bill",col="time",kde=True) # giống histogram nhưng thêm được kde 
```

#  Facet Grib 

* Facet Grid cho phép bạn dễ dàng tạo một lưới của các biểu đồ, nơi mỗi ô trong lưới hiển thị một phân loại của dữ liệu. Điều này giúp bạn so sánh phân phối của dữ liệu giữa các nhóm hoặc phân loại khác nhau.

```python
tips_df.head()
```

```python
tips_fg=sns.FacetGrid(data=tips_df,row="smoker",col="time") # Create a class instance of Facet grid class 
tips_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
new_fg=sns.FacetGrid(data=tips_df,col="sex",
                     hue="smoker",palette="Set2",
                    col_order=["Female","Male"], # col_oder ưu tiên giá trị nào xếp trước 
                    height=4,aspect=1.4) # chiều cao và lỉ lệ giữa rộng và cao 
new_fg.map(sns.scatterplot,"total_bill","tip",s=100,edgecolor='b',alpha=0.7) # 
```

#  Pair Plots


* Kiểm tra tương quan: Pair plots cho phép bạn xem xét sự tương quan giữa các biến. Các điểm dữ liệu trong biểu đồ scatter plot có thể hiển thị một hình dáng chung (nếu có mối quan hệ tuyến tính) hoặc một phân phối tập trung quanh một đường (nếu có mối quan hệ phi tuyến tính).

* Phân phối biến: Pair plots cho phép bạn xem xét phân phối của từng biến riêng lẻ thông qua histogram hoặc ước tính mật độ hạt nhân.

* Phân loại dữ liệu: Bạn có thể sử dụng pair plots để phân loại dữ liệu bằng màu sắc hoặc kích thước điểm, điều này giúp bạn kiểm tra tương quan và phân phối dữ liệu dựa trên các biến phân loại.

```python
sns.pairplot(data=penguins, hue="species") # giống joinplot nhưng nó sẽ hiển thị hết tương quan  tất cả biến 
```

# Project: Nutrition Data Visualisation on 80 cereal products
* Link: https://www.kaggle.com/crawford/80-cereals

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
file_path=r"C:\Users\sonki\test python\seaborn\data\cereal.csv"
cereal_df = pd.read_csv(file_path)
```

```python
cereal_df.head()
```

```python
cereal_df.info()
```

```python
#irrelevant fields

fields = ['shelf', 'weight', 'cups', 'rating']
cereal_df_new = cereal_df.drop(fields, axis=1)
cereal_df_new.head()
```

```python
#corr()

cereal_corr = cereal_df_new.corr(numeric_only=True) # numeric_only=True chỉ hiện thị các tương quan số 
cereal_corr 
```

* Ma trận tương quan thường có dạng bảng vuông, với các biến ở cả hai hàng và cột. Giá trị trên đường chéo chính của ma trận thường là 1, vì mỗi biến tương quan hoàn toàn với chính nó. Các giá trị nằm ngoài đường chéo chính thể hiện mức độ tương quan giữa các biến. Mức độ tương quan được biểu thị bằng các số trong khoảng -1 đến 1, trong đó:

* 1: Tương quan hoàn toàn dương (tăng cùng nhau).
* 0: Không có tương quan tuyến tính.
* -1: Tương quan hoàn toàn âm (giảm cùng nhau).

```python
# ones_like có thể xây dựng một ma trận booleans (True, False) với hình dạng giống như dữ liệu của chúng ta
ones_corr = np.ones_like(cereal_corr, dtype=bool)
ones_corr 
```

```python
ones_corr.shape, cereal_corr.shape 
```

```python
# NP's Triu: Chỉ trả về ma trận tam giác trên
mask = np.triu(ones_corr)
mask # vì mà trận đối xứng nên chỉ cần thông tin một nửa 
```

```python
sns.heatmap(data=cereal_corr,mask=mask)
```

```python
adjusted_mask = mask[1:, :-1]
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
sns.heatmap(data=adjusted_cereal_corr,mask=adjusted_mask)
```

```python
fig, ax = plt.subplots(figsize=(10,8)) # chỉ độ dài bảng 


#Phương pháp đó sử dụng màu HUSL, vì vậy bạn cần màu sắc, độ bão hòa và độ sáng. 
#I đã sử dụng hsluv.org để chọn màu của biểu đồ này.
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True) # chọn loại màu 

sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,  # vẽ dạng heatmap 
            annot=True, annot_kws={"fontsize":13}, fmt=".2f", cmap=cmap, # annot hiện thị thông số , kích thước , số chữ số thập phân 
            vmin=-1, vmax=1, # tạo colorbar bảng màu của cmap
            linecolor='white', linewidths=0.5) # tạo đường sọc trắng 

yticks = [i.upper() for i in adjusted_cereal_corr.index]  # chuyển tất cả viết hoa bằng upper 
xticks = [i.upper() for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks, rotation=0, fontsize=13)
ax.set_xticklabels(xticks, rotation=90, fontsize=13)
title = 'CORRELATION MATRIX\nSAMPLED CEREALS COMPOSITION\n'
ax.set_title(title, loc='left', fontsize=18)
```
